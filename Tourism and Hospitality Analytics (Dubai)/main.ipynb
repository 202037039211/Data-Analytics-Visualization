```python
import pandas as pd

df = pd.read_csv('Divs_Visitors_Demographic_2024-04-26_10-39-44.csv')

df.head()
```

```python
df.isnull().sum()

df.info()

df.describe()
```

```python
yearly_trends = df.groupby('year')['length_of_stay_days'].sum()

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
yearly_trends.plot(kind='bar')
plt.title('Seasonal Trends in Length of Stay')
plt.xlabel('Year')
plt.ylabel('Total Length of Stay (Days)')
plt.show()
```

```python
gender_distribution = df[['gender_male', 'gender_female']].sum()

gender_distribution.plot(kind='bar', figsize=(10,6))
plt.title('Visitor Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Visitors')
plt.show()
```

```python
purpose_distribution = df[['divs_visit_purpose_leisure', 'divs_visit_purpose_business', 
                           'divs_visit_purpose_independent_business', 'divs_visit_purpose_mice', 
                           'divs_visit_purpose_vfr']].sum()

purpose_distribution.plot(kind='bar', figsize=(10,6))
plt.title('Visitor Purpose Distribution')
plt.xlabel('Purpose of Visit')
plt.ylabel('Number of Visitors')
plt.show()
```

```python
df.to_csv('Cleaned_Tourism_Data.csv', index=False)
```
